# **Análisis de la competencia e inteligencia de mercado**

Analizaremos el conjunto de datos de bestsellers de Amazon para obtener inteligencia de mercado y realizar un análisis de la competencia. Este conjunto de datos proporciona información sobre los libros más vendidos en Amazon del 2009 al 2019.

# **Paso 1: Descargar y cargar el conjunto de datos**

En este paso, usamos el comando wget para descargar el conjunto de datos desde una URL especificada. wget es un comando de línea de comandos libre que puede usarse para descargar contenido desde la web. La opción -O nos permite especificar el nombre del archivo de salida. Luego, usamos la función read_csv de pandas para cargar el archivo CSV descargado en un DataFrame de pandas.

```python
!wget https://raw.githubusercontent.com/FelipeOrtiz-Clavijo/TwAn/main/bestsellers%20with%20categories.csv -O bestsellers_with_categories.csv

import pandas as pd

data = pd.read_csv('bestsellers_with_categories.csv')

```

# **Paso 2: Análisis exploratorio de datos**

En este paso, realizamos un análisis exploratorio inicial de los datos. Usamos la función head para ver las primeras filas del conjunto de datos, lo que nos da una idea de los campos disponibles. Luego, usamos describe para obtener un resumen estadístico de las columnas numéricas (conteo, media, desviación estándar, mínimo, 25%, 50%, 75% y máximo). Finalmente, usamos isnull().sum() para comprobar si hay valores nulos en el conjunto de datos.


```python
# Ver las primeras filas del conjunto de datos
print(data.head())

# Resumen estadístico
print(data.describe())

# Comprobar valores nulos
print(data.isnull().sum())
```

# **Paso 3: Visualización de los datos**

En este paso, utilizamos matplotlib y seaborn para visualizar nuestros datos. Creamos un gráfico de barras que muestra el conteo de libros por género, un histograma que muestra la distribución de las calificaciones de los usuarios y un gráfico de barras que muestra los 10 autores con más libros en la lista de bestsellers.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Conteo de libros por género
sns.countplot(x='Genre', data=data)
plt.show()

# Distribución de las calificaciones de los usuarios
sns.histplot(data=data, x='User Rating', kde=True)
plt.show()

# Los 10 autores con más libros en la lista de bestsellers
top_authors = data['Author'].value_counts().head(10)
top_authors.plot(kind='bar')
plt.show()
```

# **Paso 4: Análisis de la competencia**

En el último paso, realizamos un análisis de la competencia al identificar los 10 libros más revisados y los 10 libros con mejor calificación. Usamos nlargest para obtener las 10 filas con el mayor número de revisiones y creamos un gráfico de barras para visualizarlo. Luego, filtramos los libros que tienen la calificación máxima de usuario y mostramos los primeros 10 en otro gráfico de barras.

```python
# Los 10 libros más revisados
top_reviewed_books = data.nlargest(10, 'Reviews')
sns.barplot(x='Reviews', y='Name', data=top_reviewed_books, orient='h')
plt.show()

# Los 10 libros con mejor calificación
top_rated_books = data[data['User Rating'] == data['User Rating'].max()].head(10)
sns.barplot(x='User Rating', y='Name', data=top_rated_books, orient='h')
plt.show()
```
